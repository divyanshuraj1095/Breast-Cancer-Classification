### Binary Classification (0 --> cancer, 1 --> no cancer)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
breast_cancer_dataset = pd.read_csv('breastcancer.csv')
```

### Data Preprocessing

```python
breast_cancer_dataset.replace({'diagnosis':{'M': 0, 'B' : 1}},inplace = True)
```

```python
breast_cancer_dataset.head()
```

```python
breast_cancer_dataset = breast_cancer_dataset.drop(['Unnamed: 32','id'],axis =1)
```

```python
breast_cancer_dataset.describe()
```

```python
breast_cancer_dataset['diagnosis'].value_counts()
```

## 1 --> Benign
## 0 -->Malignant

```python
breast_cancer_dataset.groupby('diagnosis').mean()
```

### Separating the data

```python
X = breast_cancer_dataset.drop(['diagnosis'],axis=1)
Y = breast_cancer_dataset['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

### Splitting the data training and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=2)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, Y_train)
```

### Model valuation : accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print("accuracy_score on training data: ",training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print("accuracy_score on test data: ",test_data_accuracy)
```
